```python
import pandas as pd
```

```python
dane = pd.read_csv('C:\Users\SC60VU\Desktop\klasyfikacja.csv', delimiter= ';'
                   ,names = ('NrKontraktu', 'TypKontraktu', 'TypUmowy', 'UzytkownikRejestracja'
                             , 'TypRoszczenia', 'MiesiaceOdRCD', 'StatusKomponentu'
                             , 'Wyplata', 'UzytkownikDecyzja', 'CzyClean', 'WiekUbez'
                             , 'PlecUbez', 'MarryUbez', 'VIP'))
dane = dane.drop(columns=['NrKontraktu','StatusKomponentu']) #usuwam kolumne
dl1 = len(dane) 
dane.dropna(inplace=True)
print('{0} (tj. {1:.0%}) obserwacji zostalo usuniętych z powodu braku danych').format(dl1-len(dane),(dl1-len(dane))*1.0/len(dane))

dane.head()
```

```python
pd.get_dummies(dane['TypRoszczenia'], prefix='TypRoszczenia').head(1)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
dane = dane.join(pd.get_dummies(dane['TypRoszczenia'], prefix='TypRoszczenia'))
dane = dane.join(pd.get_dummies(dane['TypKontraktu'], prefix='TypKontraktu'))
dane = dane.join(pd.get_dummies(dane['TypUmowy'], prefix='TypUmowy'))
dane = dane.join(pd.get_dummies(dane['UzytkownikDecyzja'], prefix='UzytkownikDecyzja'))
dane = dane.join(pd.get_dummies(dane['MarryUbez'], prefix='MarryUbez'))
dane = dane.drop(columns = ['UzytkownikRejestracja','TypRoszczenia'
                     ,'TypUmowy','TypKontraktu','UzytkownikDecyzja','MarryUbez'])
```

```python
Y = dane['Wyplata']
X = dane.drop(columns = ['Wyplata'])
```

```python
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
Y_pred = dt.predict(X_test)
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
#recall- czulosc
#precision- precyzja
```

```python
#Zwiększenie minimalnych wartości zdecydowanie poprawiło model
dt1 = DecisionTreeClassifier(min_samples_leaf= 15, criterion = 'gini', min_weight_fraction_leaf = 0.01)
dt1.fit(X_train, Y_train)
Y_pred = dt1.predict(X_test)
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
